## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#Code to make your dataset analysis ready.
# Create a copy of dataset
df_clean = df.copy()

# Drop column having almost all values missing
df_clean.drop(columns=['connected_handling_time'], inplace=True)

# Fill missing categorical values
df_clean['Customer Remarks'] = df_clean['Customer Remarks'].fillna('No Remarks')
df_clean['Customer_City'] = df_clean['Customer_City'].fillna('Unknown')
df_clean['Product_category'] = df_clean['Product_category'].fillna('Unknown')

# Fill missing numerical values
df_clean['Item_price'] = df_clean['Item_price'].fillna(df_clean['Item_price'].median())

print("Data Cleaning Completed")
print("Shape after cleaning:", df_clean.shape)
````

### What all manipulations have you done and insights you found?

Data wrangling was performed to prepare the dataset for analysis. The connected_handling_time column was removed because it contained almost all missing values. Missing values in Customer Remarks, Customer_City, and Product_category were replaced with suitable placeholder values. Missing values in Item_price were filled using the median value to preserve the data distribution. No duplicate records were found in the dataset. These preprocessing steps improved data quality and made the dataset suitable for visualization, statistical analysis, and machine learning model development.
